```python
import sys
sys.path.append("../../src")
import numpy as np
import utils.plot_functions as pf
import os
import torch
import matplotlib.pyplot as plt
os.chdir("../..")
os.getcwd()
```

```python
path_to_data = "data/processed/ESC50"
resolutions = ["train_498x128", "train_128x128", "train_64x64", "train_32x32"]
res_titles = [r"498 $\times$ 128", r"128$\times$128", r"64$\times$64", r"32$\times$32"]
file_name = "3-87936-A-46.pt"
i = 0

pf.set_style()
fig, axes = plt.subplots(1,4, figsize=(pf.set_size("project", fraction=1.2, height_ratio=0.5)))
for res, ax in zip(resolutions, axes.ravel()):
    spec_path = os.path.join(*[path_to_data, res, file_name])
    spec = torch.load(spec_path).T
    im = ax.imshow(spec, origin='lower', aspect='auto', cmap="viridis")
    ax.set_title(res_titles[i])
    ax.grid(False)
    i += 1
fig.suptitle(f'Log mel-scale spectrogram of "{file_name[:-3]}.wav"')
fig.text(0.5, -0.005, 'Frames', ha='center')
fig.text(-0.005, 0.4, 'Mel bins', va='center', rotation='vertical')
plt.tight_layout()
plt.savefig("docs/Filterbank_plots/log-mel_resolutions.png", bbox_inches = 'tight')
plt.savefig("docs/Filterbank_plots/log-mel_resolutions.eps", bbox_inches = 'tight')
```
